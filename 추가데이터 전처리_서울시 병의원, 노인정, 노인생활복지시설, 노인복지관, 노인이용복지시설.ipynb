### 노인정

```python
# 필요한 데이터만 추출
condition = (file.CLSS4_NM == '노인정') 

노인정 = file.loc[condition, ['STORE_NM','CLSS4_NM', 'ADDRESS', 'HOUS_ID','CTY_CD','X_AXIS','Y_AXIS']]
노인정
```

```python
서울노인정 = 노인정[노인정['ADDRESS'].str.contains('서울')] 
서울노인정
```

```python
# 파일 내보내기
서울노인정.to_excel('/Users/zoe/Downloads/서울_노인정.xlsx')
```
